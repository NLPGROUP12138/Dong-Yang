```python
# 更改头部信息，模拟浏览器去访问网站
kv = {'user-agent':'Mozilla/5.0'}
url = "https://www.amazon.cn/gp/product/B01M8L5Z3Y"
r = requests.get(url,headers = kv)
```

```python
r.status_code
```

```python
import requests
url = "https://www.amazon.cn/gp/product/B01M8L5Z3Y"
try:
    kv = {'user-agent':'Mozilla\5.0'}
    r = requests.get(url, headers=kv)
    r.raise_for_status()
    r.encoding = r.apparent_encoding
    print(r.text[1000:2000])
except:
    print("爬取失败！")
```

```python
# 构造键值对，表明要搜索的关键词
k = {'user-agent':'Mozilla/5.0'}
kv = {'q':'Python'}
r = requests.get("http://www.so.com/s",headers=k,params = kv)
```

```python
import requests
url = "http://www.baidu.com/s"
keyword = "Python"
try:
    kv = {'wd':keyword}
    r = requests.get(url, params=kv)
    print(r.request.url)
    r.raise_for_status()
    print(len(r.text))
except:
    print("爬取失败！")
```
